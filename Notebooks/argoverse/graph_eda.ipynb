### Some visualizations and analysis

```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import os
from glob import glob
from tqdm import tqdm
```

```python
#argoverse_graphs = os.listdir("/home/tmuehlen/repos/graph_coverage/actor_graphs/argoverse")
argoverse_graphs = glob("/home/tmuehlen/repos/graph_coverage/data/argoverse_actor_and_map/*actor_graph.pkl")
print(len(argoverse_graphs))

#carla_graphs = os.listdir("/home/tmuehlen/repos/graph_coverage/actor_graphs/carla")
#print(len(carla_graphs))
```

```python

argoverse_scns = [scn.split("/")[-1].split("_")[0] for scn in argoverse_graphs]
scn = argoverse_scns[0]
```

```python

with open(f"/home/tmuehlen/repos/graph_coverage/actor_graphs/argoverse/{scn}_actor_graph.pkl", "rb") as f:
    actor_graph = pickle.load(f)

#actor_graph.timestamps
```

```python
#actor_graph.visualize_actor_graph(actor_graph.timestamps[0], 0)
for scn in tqdm(argoverse_scns):
    with open(f"/home/tmuehlen/repos/graph_coverage/data/argoverse_actor_and_map/{scn}_actor_graph.pkl", "rb") as f:
        actor_graph = pickle.load(f)
    timestamps = list(actor_graph.actor_graphs.keys())
    for timestamp in timestamps:
        # actor_graph.visualize_actor_graph(timestamp, 0)
        save_path = f'/home/tmuehlen/repos/graph_coverage/actor_graphs/argoverse_nx/graph_{scn.split("_actor_")[0]}_{str(timestamp).replace(".", "_")}.pkl'
        with open(save_path, "wb") as file:
            pickle.dump(actor_graph.actor_graphs[timestamp], file)
        
```
